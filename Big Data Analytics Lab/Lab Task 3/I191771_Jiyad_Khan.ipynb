## Identifying and Dropping Duplicate Data

````python
data = pd.read_csv('artwork_sample.csv')
````

````python
data
````

##### For some data pipe lines or machine learning models it's ok to have duplicates but in other cases we want to remove the rows that are complete dupliccates of each other or even rows that are partially duplicated

````python
data.drop_duplicates()
````

##### We still have all 10 rows that we started with. By default it drops if two rows are identical. Lets say we want to match on the subset of the columns we can pass a parameter to drop_duplicates called subset that takes an array of columns

````python
data.drop_duplicates(subset=['artist'])
````

##### Now we are only left with two rows because there were only two distinct artists in the mini dataset. What drop_duplicates did was to pick one of the duplicate rows and drop the rest and we can decide how it specifies that using keep parameter. So if we want to keep the first duplicate and drop the rest we can:

````python
data.drop_duplicates(subset=['artist'], keep='first')
````

````python
data.drop_duplicates(subset=['artist'], keep='last')
````

##### Or if we want to keep none of the duplicate rows we can:

````python
data.drop_duplicates(subset=['artist'], keep=False)
````

##### Because each of those artists had more than one row

````python
data.drop_duplicates(subset=['artist'], keep='first', inplace=True)
````
