### Filtering Data With str.contains

````python
data = pd.read_csv('artwork_sample.csv')
````

````python
data.head()
````

##### Lets say we wand to focus on medium column above and we want to fileter some of the values

````python
data.medium
````

````python
type(data.medium)
````

##### Since this is a series we can use a class of methods on it which are called string methods we can access them by calling method str on series

````python
data.medium.str
````

##### we can see that it is refencing StringMethods from pandas one of those string methods is called contains

````python
data.medium.str.contains('Graphite')
````

##### Remember to sinpily think of it as (row logic) before the comma and (column logic) after the comma

````python
data.loc[data.medium.str.contains('Graphite'), ['artist','medium']]
````

##### We got three works where Blake Robert uses Graphite

````python
data.head()
````

##### It seems like we missed the first one which is due to case sensitive nature of contains

````python
data.loc[data.medium.str.contains('Graphite', case=False), ['artist','medium']]
````

##### We  can also use regex with contains to get the case insensitive results

````python
data.loc[data.medium.str.contains('(?i)Graphite', regex=True), ['artist','medium']]
````

##### Because we are using loc to filter data here and each contains method will return a true of false for each row we can combine them with simple and's and or's like:

````python
data.loc[data.medium.str.contains('(?i)Graphite', regex=True) | data.medium.str.contains('(?i)Line', regex=True), ['artist','medium']]
````

##### This way we get all those rows where medium contains Graphite or Line

##### Another easier way of doing this is regex combined with case we set regex =True because of regex char `|` in graphite | line
##### More concise form:

````python
data.loc[data.medium.str.contains('graphite|line', case=False, regex=True), ['artist','medium']]
````
